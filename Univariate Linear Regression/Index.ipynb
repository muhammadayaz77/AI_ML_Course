```python
import numpy as np
import matplotlib.pyplot as plt
```

<h1>Load our data set</h1>

```python
x_train = np.array([1.0,2.0]) # features 
y_train = np.array([300.0,5.0]) # target value
```

<h1>Cost Function</h1>


```python
def compute_cost(x,y,w,b):
  m = x.shape[0]
  cost = 0

  for i in range(m):
    f_wb = w * x[i] + b
    cost += (f_wb - y[i])**2
  total_cost = 1 / 2 * m * cost
  return total_cost
```

<h1>Gradient Decent</h1>

```python
def comput_gradient(x,y,w,b):
  m = x.shape[0]
  dj_dw = 0
  dj_db = 0
  for i in range(m):
    f_wb = w * x[i] + b
    dj_dw_i = (f_wb - y[i]) * x[i]
    dj_db_i = (f_wb - y[i])
    dj_dw += dj_dw_i
    dj_db += dj_db_i

  dj_dw = dj_dw / m
  dj_db = dj_db / m
  return dj_dw,dj_db
```

```python
# Gradient descent function using your original functions
def gradient_descent(x, y, w_init, b_init, learning_rate, iterations):
    w = w_init
    b = b_init
    cost_history = []
    
    for i in range(iterations):
        # Compute gradient using your original function
        dj_dw, dj_db = comput_gradient(x, y, w, b)
        
        # Update parameters
        w = w - learning_rate * dj_dw
        b = b - learning_rate * dj_db
        
        # Compute cost using your original function
        cost = compute_cost(x, y, w, b)
        cost_history.append(cost)
        
        # Print progress
        if i % 100 == 0 or i == iterations-1:
            print(f"Iteration {i:4}: Cost {cost:.4f}, w={w:.3f}, b={b:.3f}")
    
    return w, b, cost_history

# Generate sample data
np.random.seed(42)
x = np.array([i for i in range(10)])
y = 2 * x + 3 + np.random.normal(0, 1, 10)  # y = 2x + 3 + noise

# Run gradient descent
w_final, b_final, costs = gradient_descent(x, y, w_init=0, b_init=0, 
                                          learning_rate=0.01, iterations=1000)

# Plot results
plt.figure(figsize=(12, 4))

# Plot 1: Data and fitted line
plt.subplot(1, 2, 1)
plt.scatter(x, y, label='Training Data')
plt.plot(x, w_final*x + b_final, 'r-', label=f'Fitted line: y={w_final:.2f}x + {b_final:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()

# Plot 2: Cost history
plt.subplot(1, 2, 2)
plt.plot(costs)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Cost Function During Training')
plt.tight_layout()
plt.show()

print(f"\nFinal parameters: w = {w_final:.4f}, b = {b_final:.4f}")
```
